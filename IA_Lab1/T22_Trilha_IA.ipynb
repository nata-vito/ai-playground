```python
class Animal:

  def __init__(self,tamanho=None,peso=None,idade=None):
    self.tamanho = tamanho
    self.peso = peso
    self.idade = idade

  def fazerbarulho(self):
    print("ahhhhhhhhhh")

  def comer(self):
    print("crac crac")

  def dormir(self):
    print("ron ron ron")

  def vaguear(self):
    print("tac tac tac")

class Felino(Animal):
  def vaguear(self):
    print("shhh shhhh shhh")

class Canino(Animal):
  def vaguear(self):
    print("tralala tralala tralala")

class Hipopotamo(Animal):
  def vaguear(self):
    print("xuaa xuaa xuuaa")

class Leao(Felino):
  def fazerbarulho(self):
    print("gruuuu")

  def comer(self):
    print("splap slap")
```

### random, randn, rand, randint, seed, normal

```python
# valores aleatorio entre zero e um
np.random.rand(4).reshape(2,2)
# amostra aleatoria de uma PDF -> gaussiana -> distribuição normal padrão (mean=0|var=1)
np.random.randn(4).reshape(2,2)
# amostra aleatoria de uma PDF - > mean e variancia customizaveis
# desvio padrão * np.random.randn(4).reshape(2,2) + media
std_v = 5
mean_v =4

ar1 = std_v * np.random.randn(4).reshape(2,2) + mean_v
print(ar1)
print("----------------")
# amostra aleatoria de uma PDF
ar2=np.random.normal(loc=mean_v,scale=std_v,size=(2,2))
print(ar2)
```

```python
# seed -> deixar a sua amostra aleatoria desterministica

std_v = 5
mean_v =4

np.random.seed(16)
ar1 = std_v * np.random.randn(4).reshape(2,2) + mean_v
print(ar1)
print("----------------")
# amostra aleatoria de uma PDF

np.random.seed(16)
ar2=np.random.normal(loc=mean_v,scale=std_v,size=(2,2))
print(ar2)
```

### Merge
image.png

```python
#merge
# key <- chave poderia ser qualquer coisa

left = pd.DataFrame({'key': ['K1', 'K4', 'K2', 'K3'],
'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
print(left)
print("---------")
print(right)
```

```python
# unir os df utilizando a 'key' como referencia -> mantendo a key do left
# pd.merge(df1, df2, como unir, referencia)
pd.merge(left=left, right=right, how='left', on='key')
```

```python
# unir os df utilizando a 'key' como referencia -> mantendo a key do right
# pd.merge(df1, df2, como unir, referencia)
pd.merge(left=left, right=right, how='right', on='key')
```

```python
# unir os df utilizando a 'key' como referencia -> mantendo a key iguais
# pd.merge(df1, df2, como unir, referencia)
pd.merge(left=left, right=right, how='inner', on='key')
```

```python
# unir os df utilizando a 'key' como referencia -> mantendo todas keys
# pd.merge(df1, df2, como unir, referencia)
pd.merge(left=left, right=right, how='outer', on='key')
```

```python
import matplotlib.pyplot as plt # abreviação tradicional
import numpy as np
```

```python
x = np.linspace(0,7,50)
y = x ** 2
```

```python
# dado continuo
# apresentação básica

plt.plot(x,y)
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Meu primeiro grafico | Matplotlib")

plt.show()
```

```python
# customizações aparência
plt.plot(x,y, color="#008000",marker='*',linestyle='--')
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Meu primeiro grafico | Matplotlib")

plt.show()
```

```python
# customizações aparência - Resumida
plt.plot(x,y, 'mo-.')
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Meu primeiro grafico | Matplotlib")

plt.show()
```

```python
# apresentação dados multiplos em um mesmo eixo
plt.plot(x,y, 'mo-.')
plt.plot(x,np.sin(x), 'b:')
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Meu primeiro grafico | Matplotlib")

plt.show()
```

```python
# apresentação dados multiplos em um mesmo eixo
plt.plot(x,y, 'r-.',x,np.sin(x), 'b:')
#plt.plot(x,np.sin(x), 'b:')
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Meu primeiro grafico | Matplotlib")

plt.show()
```

```python
# Legenda  | Latex -> Overleaf
plt.plot(x,y, 'mo-.', label="$f(x)= x^2$")
plt.plot(x,np.sin(x), 'b:', label=r"$f(x)=sin(\theta)$")
plt.plot(x,np.cos(x), 'g--', label=r"$f(x)=cos(\theta)$")
plt.legend()

plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Meu primeiro grafico | Matplotlib")

plt.show()
```

```python
# Plotagem em diferentes eixos | Subplot
# plt.subplot(linha,coluna,tag)
plt.subplot(3,1,1)
plt.plot(x,y, 'mo-.', label="$f(x)= x^2$")
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Primeiro grafico | Matplotlib")
plt.legend()

plt.subplot(3,1,2)
plt.plot(x,np.sin(x), 'b:', label=r"$f(x)=sin(\theta)$")
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Sugundo grafico | Matplotlib")
plt.legend()

plt.subplot(3,1,3)
plt.plot(x,np.cos(x), 'g--', label=r"$f(x)=cos(\theta)$")
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.title("Terceiro grafico | Matplotlib")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Instância do objeto figura
fig = plt.figure()
# Adicionando eixo
ax1= fig.add_axes([0.1,0.1,0.9,0.9])
ax2= fig.add_axes([0.2,0.6,0.4,0.3]) # [pading -- esq-> direita, pading -- baixo-> cima, largura,altura ] 0 -> 1

# plotar nos eixos

ax1.plot(x,y,"r")
ax1.set_xlabel("Eixo X")
ax1.set_ylabel("Eixo Y")
ax1.set_title("Figure | Fora")

ax2.plot(x,np.cos(x),"b", label=r'$f(x)= cos(\theta)$')
ax2.plot(x,y, 'm-.', label="$f(x)= x^2$")
ax2.legend()
ax2.set_xlabel("Eixo X")
ax2.set_ylabel("Eixo Y")
ax2.set_title("Figure | Dentro")

fig.show()
```
